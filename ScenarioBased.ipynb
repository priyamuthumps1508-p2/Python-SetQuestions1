```python
#6.to check if a student has passed a subject by scoring at least 40 marks.
```

```python
marks = list(map(int, input("Enter 5 subject marks separated by space: ").split()))

print("Marks:", marks)

failed = False

for temp in marks:
    if temp < 40:
        failed = True
        break

if failed:
    print("Failed: One or more subjects got less than 40 marks")
else:
    print("Pass: All subjects scored more than 40")
```

```python
#9.find given year is lear year or not
```

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
#10.to extract and return only the even numbers from a list.
```

```python
list_num = [10, 1, 3, 20, 14]

even_numbers = []
odd_numbers = []

for temp in list_num:
    if temp % 2 == 0:
        even_numbers.append(temp)
    else:
        odd_numbers.append(temp)

print("Even numbers from the list:", even_numbers)
print("Odd numbers from the list:", odd_numbers)
```
